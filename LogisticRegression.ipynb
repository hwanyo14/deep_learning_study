#**Perceptron: 퍼셉트론**
>이진 분류에서 최적의 가중치를 학습하는 알고리즘

하나의 뉴런(유닛)에 n개의 입력 데이터가 있을 때,   
###$$
  w_1x_1 + w_2x_2 + ... + w_nx_n + b = z\\
  z = \displaystyle\sum_{i=1}^{n}w_ix_i
$$


연산결과 z를 step function에 넣어 이진 결과값을 얻어낸다.

#**Adaline: 아달린**
퍼셉트론을 개선한 적응형 선형 뉴런   

선형 함수의 결과를 학습에 이용하고, 계단함수의 결과는 예측에만 사용

#**Logistic Regression: 로지스틱 회귀**
선형 함수를 통과시켜 얻은 z를 활성화 함수(activation function)에 통과시켜 a값을 얻는다.   
a을 학습에 이용하고, 얻은 a를 임계 함수에 통과시켜 예측값을 얻는다. 

##**Activation Function: 활성화 함수**
활성화 함수는 보통 비선형 함수를 사용한다. 다음은 대표적인 예이다.
$$
  p = \frac{1}{1 + e^{-z}}
$$

#**Sigmoid Function: 시그모이드 함수**
시그모이드 함수는 뉴런의 선형함수를 통과한 z를 입력받는다.
$$
  z = b + \displaystyle\sum_{i=1}^{n}w_ix_i \quad (선형함수)
$$

z를 입력받아 0 ~ 1 사이의 확률을 갖는 a를 도출해낸다.

###$$
  p = \frac{1}{1 + e^{-z}} \quad (시그모이드 함수)
$$


```python
import matplotlib.pyplot as plt
import numpy as np

z = np.arange(-10, 10)
p = 1 / (1 + np.exp(-z))
plt.plot(p)
plt.xlabel("z")
plt.ylabel("1 / (1 + exp(-z))")
plt.show()
```

#**Logistic Loss Function: 로지스틱 손실함수**

 - 선형회귀는 **타깃과 예측값의 오차 제곱을 최소화**하는 가중치와 절편을 찾는 것이 목표
 - 로지스틱 회귀는 **분류된 데이터의 비율**을 높이는 것이 목표

###$$
  L = -(ylog(a) + (1 - y)log(1 - a)) \quad (a: 활성화 함수 결과값, y: 타깃)
 $$
|||
|-----|------|
|y가 1인 경우(양성)|$-log(a)$|
|y가 0인 경우(음성)|$-log(1 - a)$|

위의 식의 값을 최소로 만들다 보면 a의 값이 목표치에 근사하는 것을 볼 수 있다.   
y = 1(양성)인 경우 로지스틱 손실함수를 최소화 하는 a는 1에 근사하고,   
y = 0(음성)인 경우 로지스틱 손실함수를 최소화 하는 a는 0에 근사한다.   
<br>
이 로지스틱 손실함수를 최소화하기 위해 미분을 한다.

###$$
\begin{aligned}
  \frac{∂}{∂w_i}L &= -(y - a)x_i\\
  \frac{∂}{∂b}L &= -(y - a)1
\end{aligned}
$$

제곱 오차 미분식에서 y_hat이 a로 바뀐 형태이다.

여기서 오차율(err)는 두 미분식의 공통된 부분인 $-(y - a)$라고 볼 수 있다.
