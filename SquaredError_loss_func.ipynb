#**Loss Function: 손실함수**
**경사하강법**을 기술적으로 표현하면,   
>"어떤 손실함수가 정의되었을 때, 손실함수의 값이 최소가 되는 지점을 찾아가는 방법"

앞에서 진행했던 방법(오차를 변화율에 곱하여 가중치에 더하기)은 '제곱오차'라는 손실함수를 미분한 것과 같다.

#**Squared Error: 제곱 오차**
제곱 오차는 타깃값에서 예측값을 뺀 것에 제곱을 한 것이다.   
>$$SE = (y - y\_hat)^2$$


###**제곱 오차를 가중치 w에 대해 미분**

###$$
  \frac{∂SE}{∂w} = 
  \frac{∂}{∂w}(y - y\_hat)^2 = 
  2(y - y\_hat)(-\frac{∂}{∂w}y\_hat) = 
  2(y - y\_hat)(-x) = 
  -2(y - y\_hat)x
$$

여기서 편미분 될 w는 y_hat에 존재하므로,   
y_hat = w * x + b를 미분하면 x만 남는다.
<br><br>
따라서,
###$$
  \frac{∂SE}{∂w} = -2(y - y\_hat)x
$$
<br><br>
여기서 제곱오차 함수를 $(y - y\_hat)^2$가 아니라 $\frac{1}{2}(y - y\_hat)^2$로 정의했다면,   
미분식이 $-(y - y\_hat)x$로 깔끔하게 표현된다.   
손실함수에 상수를 곱하거나 나누어도 최종 모델의 가중치나 절편에 영향을 주지 않으므로 일반적으로 사용되는 $-(y - y\_hat)x$식을 사용한다.

###**제곱 오차의 변화율을 이용해 가중치 업데이트**
가중치 업데이트를 위해 w에서 변화율을 뺀다.   
(손실함수를 낮은 쪽으로 이동하기 위해 뺌)

###$$
  w = w - \frac{∂SE}{∂w} = w + (y - y\_hat)x
$$

```python
y_hat = x_i * w + b
err = y - y_hat
w += x_i * err    # 앞서 진행했던 오차율을 변화량과 곱하여 더하는 방식과 같다.
```

###**제곱 오차를 절편에 대하여 미분**
###$$
  \frac{∂SE}{∂b} = \frac{∂}{∂b}\frac{1}{2}(y - y\_hat)^2 = 
  (y - y\_hat)(-\frac{∂}{∂b}y\_hat) = 
  (y - y\_hat)(-1) = -(y - y\_hat)1
$$
<br><br>
마찬가지로 절편 b에서 변화율을 뺀다.

$$
  b = b - \frac{∂SE}{∂b} = b + (y - y\_hat)
$$

```python
err = y_i - y_hat
b += 1 * err  # 앞서 진행했던 식과 완전히 일치
```
